```python
df.drop(['Surname','CustomerId','RowNumber'],axis=1,inplace=True)
```

```python
df
```

```python
x=df.iloc[:,:-1]
y=df['Exited']
```

```python
x
```

```python
y
```

```python
Gender=pd.get_dummies(x['Gender'])
Geography=pd.get_dummies(x['Geography'])
x=x.drop(['Gender','Geography'],axis=1)
```

```python
x=pd.concat([x,Gender,Geography],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## create a ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

## Initialize the ANN

```python
classifier=Sequential()
```

## Adding input layer

```python
classifier.add(Dense(units=11,activation='relu')) ##  11=columns(input)
```

## Adding the first hidden layer


```python
classifier.add(Dense(units=7,activation='relu'))
```

## Adding the second hidden layer

```python
classifier.add(Dense(units=6,activation='relu'))
```

## Adding the output layer

```python
classifier.add(Dense(units=1,activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
```

```python
model_history=classifier.fit(x_train,y_train,validation_split=.33,batch_size=10,epochs=1000,callbacks=early_stopping)
```

## Early stopping

```python
early_stopping=tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```
